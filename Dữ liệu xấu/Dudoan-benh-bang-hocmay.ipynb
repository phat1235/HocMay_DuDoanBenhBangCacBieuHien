```python
# Điền các giá trị NaN bằng giá trị phía trên
data = df.fillna(method='ffill')
```

```python
data.head(30)
```

```python
# Xử lý tên bệnh và triệu chứng
def process_data(data):
    data_list = []
    data_name = data.replace('^','_').split('_')
    n = 1
    for names in data_name:
        if (n % 2 == 0):
            data_list.append(names)
        n += 1
    return data_list
```

```python
# Dọn dẹp dữ liệu
disease_list = []
disease_symptom_dict = defaultdict(list)
disease_symptom_count = {}
count = 0

for idx, row in data.iterrows():

        # Lấy tên bệnh
    if (row['Disease'] !="\xc2\xa0") and (row['Disease'] != ""):
        disease = row['Disease']
        disease_list = process_data(data=disease)
        count = row['Count of Disease Occurrence']

 # Lấy triệu chứng tương ứng với bệnh
    if (row['Symptom'] !="\xc2\xa0") and (row['Symptom'] != ""):
        symptom = row['Symptom']
        symptom_list = process_data(data=symptom)
        for d in disease_list:
            for s in symptom_list:
                disease_symptom_dict[d].append(s)
            disease_symptom_count[d] = count
```

```python
# Kiểm tra dữ liệu đã được xử lý đúng CHƯA
disease_symptom_dict
```

```python
# Đếm số lần xuất hiện của biểu hiện bệnh tương ứng với từng bệnh
disease_symptom_count
```

```python
# Lưu dữ liệu đã dọn dẹp vào file CSV
f = open('/content/drive/My Drive/Colab Notebooks/Check-check-test-test/dataset/cleaned_data.csv', 'w')

with f:
    writer = csv.writer(f)
    for key, val in disease_symptom_dict.items():
        for i in range(len(val)):
            writer.writerow([key, val[i], disease_symptom_count[key]])
```

```python
# One Hot Encode nhãn
onehot_encoder = OneHotEncoder(sparse_output=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```
